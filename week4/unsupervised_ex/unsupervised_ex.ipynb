```python
#Generate a random dataset with two features (x1, x2)
import numpy as np

# Set the random seed for reproducibility
np.random.seed(42)

# Define the number of samples in the dataset
num_samples = 100

# Generate random values for x1 and x2
x1 = np.random.rand(num_samples)
x2 = np.random.rand(num_samples)

# Combine x1 and x2 into a single dataset
dataset = np.column_stack((x1, x2))

# Print the first few rows of the dataset
print("Generated Dataset:")
print(dataset[:5])
```

```python
#Visualize the dataset to get an idea of what it looks like. You can use a scatter plot to visualize the data points.
import matplotlib.pyplot as plt

# Visualize the dataset
plt.scatter(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Scatter Plot of the Generated Dataset')
plt.show()
```

```python
#Implement the K-Means clustering algorithm and use scatter plot
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a numpy array from the generated dataset
data = np.column_stack((x1, x2))

# Define the number of clusters
k = 3

# Create a K-Means clustering model
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model to the data
kmeans.fit(data)

# Get the cluster labels and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Visualize the clusters
plt.scatter(x1, x2, c=cluster_labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='red', label='Centroids')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

# Create a K-Means clustering model
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model to the data
kmeans.fit(data)

# Get the inertia score
inertia = kmeans.inertia_

# Print the inertia score
print("Inertia Score:", inertia)
```

```python
from sklearn.metrics import silhouette_score

# Get the cluster labels
cluster_labels = kmeans.labels_

# Calculate the silhouette score
silhouette = silhouette_score(data, cluster_labels)

# Print the silhouette score
print("Silhouette Score:", silhouette)
```
