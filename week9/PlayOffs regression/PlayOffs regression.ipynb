```python
import numpy as np
import pandas as pd
import seaborn as s ns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PlayOffs.csv')
df
```

```python
corr=df.corr()
corr
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr,xticklabels=True,yticklabels=True,annot=True)
```

```python
x = df['Minutes_played']
y = df['PersonlFoul']
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

# Assuming you have already defined x and y from your DataFrame
x = df['Minutes_played']
y = df['PersonlFoul']

# Reshape the x array to a 2D array since scikit-learn expects a 2D input
X = np.array(x).reshape(-1, 1)

# Fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)

# Predict the y values based on the fitted model
y_pred = regressor.predict(X)

# Plot the scatter plot of the data points
plt.scatter(x, y)

# Plot the linear regression line
plt.plot(x, y_pred, color='red')

# Add labels and title to the plot
plt.xlabel('Minutes Played')
plt.ylabel('Personal Foul')
plt.title('Linear Regression')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.residplot(data=df,x='Minutes_played',y='PersonlFoul',lowess=True,line_kws=dict(color='r'))
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Fit a linear regression model on the training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict the y values for the test data
y_pred = regressor.predict(X_test)

# Plot the scatter plot of the test data points
plt.scatter(X_test, y_test)

# Plot the predicted values from the test data
plt.plot(X_test, y_pred, color='red')

# Add labels and title to the plot
plt.xlabel('Minutes Played')
plt.ylabel('Personal Foul')
plt.title('Test Model: Linear Regression')

# Display the plot
plt.show()
```

```python

plt.scatter(X_test, y_test)

# Plot the predicted values from the test data
plt.plot(X_test, y_pred, color='red')

# Add error lines between the points and the regression line
for i in range(len(X_test)):
    plt.plot([X_test[i], X_test[i]], [y_test.iloc[i], y_pred[i]], color='blue', linestyle='--')

# Add labels and title to the plot
plt.xlabel('Minutes Played')
plt.ylabel('Personal Foul')
plt.title('Test Model: Linear Regression with Error Lines')

# Display the plot
plt.show()
```
