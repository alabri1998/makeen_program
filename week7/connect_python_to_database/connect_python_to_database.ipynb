```python
#calculate average of age colomn in players database
import pyodbc

def calculate_average(table, column):
    try:
        # Establish a connection to the SQL Server database
        conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=LAPTOP-S5Q8O6V9\MSSQLS;'
                      'Database=Football Players;'
                      'Trusted_Connection=yes;')

        # Create a cursor object to execute SQL queries
        cursor = conn.cursor()

        # Prepare the SQL query to retrieve the column values
        query = f"SELECT {column} FROM {table}"

        # Execute the query and fetch all rows
        cursor.execute(query)
        rows = cursor.fetchall()

        # Calculate the average value
        total_sum = 0
        row_count = len(rows)
        for row in rows:
            value = row[0]  # Assuming the column is in the first position (index 0)
            total_sum += float(value)  # Convert the value to float if necessary

        if row_count > 0:
            average = total_sum / row_count
            return average
        else:
            return None

    except pyodbc.Error as e:
        # Handle any errors that occur during the execution
        print(f"Error calculating average: {e}")

    finally:
        # Close the cursor and the database connection
        cursor.close()
        conn.close()

# Usage example

table_name = 'players'
column_name = 'age'

average_value = calculate_average( table_name, column_name)
if average_value is not None:
    print(f"The average value of column '{column_name}' is: {average_value}")
else:
    print(f"No data found in column '{column_name}'.")
```

```python
#selects a subset of records based on specific criteria, and performs statistical calculations.
import pyodbc
import statistics

def calculate_statistics( table, column , criteria):
    try:
        # Establish a connection to the SQL Server database
        conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=LAPTOP-S5Q8O6V9\MSSQLS;'
                      'Database=Football Players;'
                      'Trusted_Connection=yes;')

        # Create a cursor object to execute SQL queries
        cursor = conn.cursor()

        # Prepare the SQL query to select the subset of records based on criteria
        query = f"SELECT {column} FROM {table} WHERE {criteria}"

        # Execute the query and fetch all rows
        cursor.execute(query)
        rows = cursor.fetchall()
        print(rows)
        # Extract the column values
        values = [float(row[0]) for row in rows]  
        
        # Calculate the mean and standard deviation
        mean = statistics.mean(values)
        std_dev = statistics.stdev(values)

        # Print the calculated statistics
        print(f"Mean of '{column}' with age > 22 is: {mean}")
        print(f"Standard deviation of '{column}' with age > 22 is: {std_dev}")

    except pyodbc.Error as e:
        # Handle any errors that occur during the execution
        print(f"Error calculating statistics: {e}")

    finally:
        # Close the cursor and the database connection
        cursor.close()
        conn.close()

# Usage example
table_name = 'players'
column_name = 'age'
criteria = 'age > 22'

calculate_statistics(table_name, column_name, criteria)
```

```python
#create new database using python
import pyodbc


def create_database(server, database_name):
    try:
        # Establish a connection to the SQL Server instance
        conn = pyodbc.connect(f"Driver={{SQL Server}};Server={server};Trusted_Connection=yes;")
        conn.autocommit = True  # Disable autocommit

        cursor = conn.cursor()

        # Create the SQL command to create the new database
        create_db_query = f"CREATE DATABASE {database_name};"

        # Execute the SQL command
        cursor.execute(create_db_query)

        # Close the cursor and connection
        cursor.close()
        conn.close()

        print(f"Database '{database_name}' created successfully!")
    except pyodbc.Error as e:
        # Handle any errors that occur during the database operation
        print("An error occurred while creating the database:")
        print(e)


# Example usage:
server_name = 'LAPTOP-S5Q8O6V9\MSSQLS'
new_database_name = 'ahmed_new'

# Call the function to create a new database
create_database(server_name, new_database_name)
```
