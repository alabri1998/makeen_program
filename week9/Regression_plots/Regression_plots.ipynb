```python
#import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
```

```python
#load dataset

mpg = sns.load_dataset('mpg')
mpg
```

```python
#Let's use lmplot() to plot the acceleration versus mpg
sns.lmplot(x="acceleration",y="mpg",data=mpg)
```

when acceleration increase mpg increase

```python
#regiression line
slope, intercept, r_value, p_value, std_err = stats.linregress(mpg['mpg'],mpg['acceleration'])
print("y="+str(np.round(slope,2))+"x+"+str(np.round(intercept,2)))
```

slope = 0.15 , y_intersept = 12.08

```python
# see the line for confidence interval = 95%
sns.lmplot("acceleration","mpg",mpg,ci=95)
```

```python
# Just like before, we can use dictionaries to edit individual parts of the plot
sns.lmplot(x="acceleration", y="mpg", data=mpg,
        scatter_kws={"marker": "o", "color": "orange"},
        line_kws={"linewidth": 4, "color": "black"});
```

```python
# WE can also check out higher-order trends
sns.lmplot("acceleration", "mpg", mpg,order=4,
scatter_kws={"marker": "o", "color": "indianred"},
line_kws={"linewidth": 1, "color": "blue"})
```

we can see that trend acceleration between 15 and 17.5 and mpg around 25

```python
# We can also not fit a regression if desired
sns.lmplot("acceleration", "mpg", mpg,fit_reg=False)
```
