```python
import requests
from bs4 import BeautifulSoup

topic = input("Enter topic: ")
# Specify the URL of the page you want to scrape
url = 'https://en.wikipedia.org/wiki/'+ topic
print(url)

# Send a GET request to fetch the page content
response = requests.get(url)

# Check the response status
if response.status_code == 200:
    # Convert the content to a Beautiful Soup object for easier processing
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the main title of the page
    title = soup.find('h1', {'id': 'firstHeading'}).text
    print('Page Title:', title)
    
    # Extract the article or main text content of the page
    content = soup.find('div', {'id': 'mw-content-text'}).text
    print('Content:', content)
else:
    print('An error occurred while fetching the page')
```
