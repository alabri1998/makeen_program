```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#generate 50 random values between 1 and 100
data = np.random.rand(50)*100
data
```

```python
#plot normal histogram
plt.hist(data)
plt.xlabel('value')
plt.ylabel('frequency')
```

```python
#when we change number of bins a shape a little bit change
plt.hist(data, bins=13)
plt.show()
```

```python
# another histogram
plt.hist(data,color='green',edgecolor='black')
```

```python
#another data
data2 = np.random.rand(50)*100
data2
```

```python
plt.hist(data,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(data2,density=True,alpha=0.5,bins=20);
```

```python
#use joint plot
sns.jointplot(data,data2)
```

```python
#Can also use hex bins for a more concise picture
#kind=scatter | reg | resid | kde | hex
p=sns.jointplot(data,data2,kind='hex')
p.set_axis_labels("Data1", "Data2")
```

```python
#Kernal Desnity Destribution to estimate histogram
sns.distplot(data)
```

```python
#Kernal Desnity Destribution to estimate histogram
sns.kdeplot(data)
```

```python
 #Kernal Desnity Destribution
sns.kdeplot(data)
#plot different KDD with different bandwidth "bw" and shaded
for i in np.arange(0.5,3,0.5):
    sns.kdeplot(data,bw=i,shade=True)
```
