### JointGrid of Sleep Duration


```python
g= sns.JointGrid(data = df,x=df.Age,y=df['Sleep Duration'])
g.plot_joint(sns.kdeplot,
             fill=True,
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```

### heatmap of dataframe


```python
import numpy as np
corr = df.corr(method = 'spearman')

#convert correlation to numpy array
mask = np.array(corr)

#to mask the repetitive value for each pair
mask[np.tril_indices_from(mask)] = False
fig, ax = plt.subplots(figsize = (15,12))
fig.set_size_inches(70,12)
sns.heatmap(corr, mask = mask, vmax = 0.9, square = True, annot = True)
```

### Linear Regression between Quality of Sleep and Stress Level

```python
from sklearn.linear_model import LinearRegression
X = df[['Quality of Sleep']]
y = df['Stress Level']
regressor = LinearRegression()
regressor.fit(X,y )
```

```python
y_pred = regressor.predict(X)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Quality of Sleep vs Stress Level')
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.show()
```
